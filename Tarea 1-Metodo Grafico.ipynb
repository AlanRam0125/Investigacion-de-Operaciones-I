### Ejercicio 1

Una compañía fabrica dos tipos de motores pequeños con combustible sólido. Con el motor A gana $ \$  3$ por motor y con el motor B gana $\$ 4$ por motor. Dispone de 80 horas por semana para producirlos y en los procesos se requieren: 4 h para A y solo 2 h para B. Sin embargo, debido a lo peligroso del material usado, son necesarias 5 h de preparación y limpieza para el motopr B y 2 h para A. El tiempo de preparación total disponible es 120 h por semana. Determinar el número de cada clase de motores a producir que dé como resultado la máxima ganancia.   

#### SOLUCIÓN:

De manera formal:

#### Variables de decisión:

Motor A = $x_{1}$  
Motor B = $x_{2}$

#### Función Objetivo:

Maximizar $3x_1 + 4x_2$

#### Restricciones:

Horas disponibles por semana para producir los motores:  
$4x_1 + 2x_2 \leq 80$  
Horas disponibles por semana para preparación y limpieza:  
$2x_1 + 5x_2 \leq 120$

No negatividad: $x_1 \geq 0$, $x_2 \geq 0$

Por lo tanto el modelo de programacion matemática para este problema es el siguiente: 

Máx Z=$3x_1 + 4x_2$  
s.a.   
$4x_1 + 2x_2 \leq 80$  
$2x_1 + 5x_2 \leq 120$  
$x_1$, $x_2 \geq 0$

Reescribiendo la primer restricción en forma estándar: 

$4x_{1} + 2x_{2} = 80$

Si $x_{1} = 0 \longrightarrow x_{2} = 40 $

Si $x_{2} = 0 \longrightarrow x_{1} = 20 $

Ahora graficando $4x_{1} + 2x_{2} = 80$

$$
\begin{align}
4x + 2y &= 80 \quad \text{(1)} \\
2x + 5y &= 120 \quad \text{(2)}
\end{align}
$$

Multiplicando por $-2$ la ecuacion (2) y sumando la ecuación 1 obtenemos:  
$$-8x_{2} = -160 \longrightarrow x_{2} = 20$$  
Sustituyendo el valor obtenido de $x_{2}$ en la ecuacion (1), obtenemos: 
$$2x_{1} = 20 \longrightarrow x_{1} = 10$$
